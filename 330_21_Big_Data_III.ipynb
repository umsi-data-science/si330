```pyspark
# This takes about 3 minutes (180s) on a single-node m4.xlarge cluster
df = spark.read.json('s3://umsi-data-science/data/yelp/review.json.gz')
```

```pyspark
df.printSchema()
```

### How many reviews are there?

```pyspark
df.count()
```

### How many positive (i.e. 4- or 5-star) reviews are there?  How many negative (i.e. 1- or 2-star) reviews are there?

```pyspark
positive_reviews = df.filter(df.stars>4).sample(False,0.001)
negative_reviews = df.filter(df.stars<2)#.sample(False,0.001)
```

### What is the average number of words in the positive reviews? 

```pyspark
pos_review_words = positive_reviews.rdd.map(lambda x: x.text.lower()) \
    .flatMap(lambda x: x.split()) \
    .filter(lambda x: x not in sw.value) \
    .map(lambda x: (x,1)) \
    .reduceByKey(lambda x,y: x+y) \
    .sortBy(lambda x: x[1], ascending=False)
```

```pyspark
print(pos_review_words.count())
```

```pyspark
positive_reviews_partitioned = df.filter(df.stars>4).sample(False,0.001).repartition(64)
```
